## Step 5: Line Chart 'prices'

### Sampling Data

```python
# Create a list holding True/False values to the test np.random.rand()<=0.7

np.random.seed(4)
dev = np.random.rand(len(df_merged)) <= 0.7
```

```python
dev
```

```python
# Store 70% of the data in dataframe big
# Store 30% of the data in dataframe small

big = df_merged[dev]
small = df_merged[~dev]
```

```python
len(df_merged)
```

```python
len(big) + len(small)
```

```python
# Create a dataframe from subset 'small' 

df_3 = small[['order_hour_of_day','prices']]
```

### Line Chart

```python
# Create Chart 

List_hour = sns.lineplot(data = df_3, x = 'order_hour_of_day',y = 'prices')
```

```python
# Export the Chart

List_hour.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'List_hour.png'))
```

## Step 6: Line Chart 'age' and 'number of dependants'
