## Creating Line Charts

### Sampling Data

```python
# Create a list holding True/False values to the test np.random.rand()<=0.7

np.random.seed(4)
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
dev
```

```python
# Store 70% of the data in dataframe big
# Store 30% of the data in dataframe small

big = ords_prods_merge[dev]
small = ords_prods_merge[~dev]
```

```python
len(ords_prods_merge)
```

```python
len(big) + len(small)
```

```python
# Create a dataframe from subset 'small' 

df_2 = small[['orders_day_of_week','prices']]
```

### Line Chart

```python
# Create a Line chart

line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Save the chart as 'List' 

List = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Check the output

List.figure
```

```python
# Export the Chart

List.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'List_prices.png'))
```
