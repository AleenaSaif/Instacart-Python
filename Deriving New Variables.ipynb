## User-defined Function

```python
# Define a function

def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range Product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range Product'
    elif row['prices'] > 15:
        return 'High-range Product'
    else:
        return 'Not enough data'
```

```python
# Create a column price_range to return the data from user_defined function price_label

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

## If-Statements with For-Loops

```python
df_ords_prods['orders_day_of_week'].value_counts(dropna = False)
```

```python
result = []

for value in df_ords_prods["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
df_ords_prods['busiest_day'] = result
```

```python
df_ords_prods['busiest_day'].value_counts(dropna = False)
```

```python
result_2 = []

for value in df_ords_prods["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Regular days")
```

```python
df_ords_prods['busiest_days'].value_counts(dropna = False)
```

## Step 4:

```python
df_ords_prods['order_hour_of_day'].value_counts(dropna = False)
```

```python
hour = []

for value in df_ords_prods["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    hour.append("Most orders")
  elif value in [3, 4, 2, 5, 1, 0, 6, 23]:
    hour.append("Fewest orders")
  else:
    hour.append("Average orders")
```

```python
hour
```

## Step 5: 

```python
df_ords_prods['busiest_period_of_day'] = hour
```

```python
df_ords_prods['busiest_period_of_day'].value_counts(dropna = False)
```
