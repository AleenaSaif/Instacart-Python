## Step 6

```python
# Create groups based on the 'user_id' column. Apply the transform() function on the 'prices' column to generate the average prices for each user.

ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Check the output

ords_prods_merge.head(10)
```

## Step 7

```python
# Create group based on the 'user_id' column. Apply the transform() function on the 'days_since_prior_order' column to generate the median order frequency for each user.

ords_prods_merge['median_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_frequency']  > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_frequency'] > 10) & (ords_prods_merge['median_frequency'] <= 20), 'order_frequency_flag'  ] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Check the output

ords_prods_merge.head(10)
```
